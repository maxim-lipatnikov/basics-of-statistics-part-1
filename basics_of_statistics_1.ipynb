## **1. Генеральная совокупность и выборка**
<img src="images/1. sampling mindmap.jpg"  width="500"/>

**Генеральная совокупность (statistical population, population, sp)** - множество всех объектов, вывод о которых мы хотим сделать.

**Примеры:**

1. Если мы врачи и лечим какую-то болезнь Х, и хотим протестировать новое лекарство для ее лечения, то все люди, 
болеющие Х, будут нашей ГС.

2. Если мы хотим исследовать влияние какого-то политического заявления на общественное мнение совершеннолетних мужчин, 
живущие в городе Х, то все совершеннолетние мужчины города Х - это наша ГС.

**Суть статистики как науки** - мы хотим сделать какие-то выводы о всей ГС, но собрать данные о всей ГС слишком сложно, 
долго и дорого (т.е. невозможно), поэтому мы берем из ГС выборку, делаем по ней какие-то выводы и обобщаем их на всю ГС. 

Для этого выборка должна быть максимально репрезентативной, т.е. должна максимально хорошо отражать свойства ГС 
(по сути, должна быть моделью ГС).

Репрезентативность выборки зависит от метода ее формирования, всего методов три:

**1. Простая случайная выборка (simple random sample)** - мы просто берем ГС и случайным образом выбираем из нее элементы. 
Чем больше элементов в нашей выборке - тем она репрезентативнее.

**Пример:** Для того чтобы протестировать курс по введению в статистику и выяснить, 
насколько хорошо с ним справятся студенты Института биоинформатики, 
случайным образом было выбрано 50 студентов института.


**2. Стратифицированная выборка (stratified sample)** - перед тем, как случайно выбрать элементы ГС, мы делим ГС на страты/группы, 
например, на мужчин и женщин. После разделения ГС на страты, из каждой страты случайным образом берем элементы.

**Пример:** Для исследования взаимосвязи риска развития заболевания и группы крови человека потенциальных участников исследования разделили на четыре группы (в соответствии с группой крови). Затем из каждой группы случайным образом извлекли по 50 человек.


**3. Групповая выборка (cluster sample)** - метод похож на стратифицированную выборку, но тут мы выделяем группы, похожие друг на друга. 
Делается это, как правило, для экономии ресурсов. Например, мы можем разделить СПБ на районы и сделать предположение о том, 
что население районов не сильно отличается друг от друга, поэтому мы можем случайным образом собрать выборку не из всех, 
а только из нескольких групп (кластеров).

**Пример:** Чтобы проверить знания школьников Санкт-Петерубрга по математике, было организовано исследование. Случайным образом было выбрано 10 школ, затем из каждой школы случайным образом было отобрано по 50 учащихся разных классов.


## **2. Типы переменных. Количественные и номинативные переменные**
<img src="images/3. central tendency mindmap.jpg"  width="300"/>

Все переменные можно разделить на две группы:
    
**1. Количественные (например, рост)**
    
- **Непрерывные** - если переменная может принимать любые значения (например, рост может быть любой между [160,190])
    
- **Дискретные** - если переменная может принимать определенные значения (например, число детей в семье может быть равно 1, 2, 3, …, но не может быть равно 3.5)

**2. Качественные (номинативные)** - это своеобразные маркеры/лейблы, которые используются для разбиения наших наблюдений на группы (например, мы можем сказать, что все женщины в нашем исследовании обозначены числом 1, а все лица мужчины обозначены числом 2)

**3. Ранговые** - это что-то среднее между количественной и номинативной переменной, которая позволяет нам расположить наблюдения по порядку, но не позволяет нам понять количественную разницу между переменными (например, места на соревнованиях по бегу - 1, 2, 3, …, но мы не знаем точно, на сколько один быстрее другого)

## **3. Меры центральной тенденции**
<img src="images/4. variability mindmap.jpg"  width="300"/>

### 3.1 Что такое распределение?

1. В случае **номинативной переменной** мы можем только посмотреть количество наблюдений в каждой группе.

2. В случае **количественной переменной** мы можем построить гистограмму частот, где можем показать, как часто значение нашей переменной встречается на том или ином промежутке значений. Гистограмма частот позволяет первично оценить распределение переменной и его форму.

### 3.2 Меры центральной тенденции (среднее, мода, медиана) 
МЦТ позволяют описать распределение одним числом.

**Мода (mode)** - это значение признака, которое встречается максимально часто. Если значение встречается несколько раз - у нас несколько мод.

**Медиана (median)** - значение признака, которое делит упорядоченную выборку пополам. Если в выборке четное кол-во элементов, то медианой будет среднее арифметическое двух чисел, которые делят выборку пополам.

**Среднее (mean, среднее арифметическое)** - сумма всех значений измеренного признака, деленная на кол-во измерений.

*Среднее выборки обозначается как $ \bar{x} $ или $ M_{x} $

*Среднее генеральной совокупности обозначается как $ M $ или $ \mu $

### 3.3 Свойства среднего значения
$$M_{x + c} = \frac{\sum_{i=1}^{n}{(x_{i} + c)}}{n} = \frac{\sum_{i=1}^{n} x_{i}}{n} + \frac{\sum_{i=1}^{n} c}{n} = M_{x} + \frac{nc}{n} = M_{x} + c$$

$$M_{x * c} = \frac{\sum_{i=1}^{n}{(x_{i} * c)}}{n} = \frac{c * \sum_{i=1}^{n} x_{i}}{n} = c * M_{x}$$

$$\sum_{i=1}^{n} (x_{i} - M_{x}) = nM_{x} - nM_{x} = 0$$

```python
# 1. Расчёт моды, медианы и среднего в numpy и scipy
import numpy as np
from scipy import stats
import pandas as pd

sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

print('mode:', stats.mode(sample))
print('median:', np.median(sample))
print('mean:', np.mean(sample))
```

```python
# 2. Расчёт моды, медианы и среднего в pandas
import pandas as pd
sample = pd.Series([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

print('mode:', sample.mode())
print('median:', sample.median())
print('mean:', sample.mean())
```

Потому что иногда они почти равны друг другу, а иногда сильно отличаются, например, как тут (красные линии - это две моды, синяя - среднее):

## **4. Меры изменчивости**


**1. Размах (range)** - разность максимального и минимального значения признака.

*Размах очень зависит от максимального и минимального значения, а значит - очень чувствителен к выбросам

**2. Дисперсия (variance)** - средний квадрат отклонений индивидуальных значений признака от их средней величины.

Дисперсия позволяет нам понять, насколько сильно в среднем каждый из элементов ГС/выборки отличается от среднего ГС/выборки.

**Для генеральной совокупности**
$$Var = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})^2}{n}$$

**Среднеквадратическое отклонение (в случае ГС)**
$$ \sigma = \sqrt{Var}$$

**Для выборки**
$$Var = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})^2}{n-1}$$

**Стандартное отклонение (в случае выборки)**
$$std = \sqrt{Var}$$

```python
# Расчитываем размах, дисперсию и стандартное отклонение с помощью numpy
import numpy as np
sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

# The name of the function comes from the acronym for ‘peak to peak’.
print(f'Range: {np.ptp(sample)}')

# ddof - Delta Degrees of Freedom
print(f'Variance: {np.var(sample, ddof=1):.2f}')
print(f'Standard deviation: {np.std(sample, ddof=1):.2f}')
```

### 4.1 Почему в формуле для ГС в знаменателе n, а в формуле для выборки n-1?
В формуле в знаменателе указано количество степеней свободы. 

Когда мы считаем дисперсию по выборке, мы хотим получить **наиболее точную оценку дисперсии ГС**.

Если мы **НЕ** вычитаем 1 в знаменателе, мы получим **смещенную оценку дисперсии ГС**, которая менее "точно" оценивает дисперсию ГС
по сравнению с **несмещенной**. 

Для получения несмещенной оценки необходимо вычесть 1 из n.

### 4.2 Почему мы берем именно КВАДРАТ отклонений значений признака от среднего?
1. Если мы оставим в числителе просто (xi - Mx), то числитель окажется равен **нулю**, т.к. это третье свойство мат. ожидания.
2. Модуль не берут потому, что модуль - не гладкая функция, т.к. в нуле у модуля имеется "излом", из-за которого у производной происходит разрыв. 

### 4.3 Свойства дисперсии

$$ Var_{x+c} = Var_x $$
$$ std_{x+c} = std_x $$

*Если мы к каждому элементу выборки прибавим константу, то форма распределения не изменится => не изменится и дисперсия

$$ Var_{x*c} = Var_x*c^2 $$
$$ std_{x*c} = std_x*c $$

*Если мы каждый элемент выборки умножим на константу, то форма распределения изменится => изменится и дисперсия 
(например, если c>1, то распределение "растянется", а если c<1 - оно "сузится")

## **6. Нормальное распределение**

**Нормальное распределение (Гауссово распределение, Normal distribution, Gaussian distribution):**
 - Унимодально
 - Симметрично
 - Отклонения наблюдений от среднего подчиняются определённому вероятностному закону ("**правилу 3-х сигм**")
 

```python
# Generating a normally distributed sample
mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Plotting the sample
plt.figure(figsize=(5,3))
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show();
```

### 6.1 Z-преобразование (стандартизация, Z-standardization)

Преобразование полученных данных, имеющих **нормальное распределение с какими-то средним и дисперсией**
в стандартную **Z-шкалу (Z-scores) со средним значением = 0 и дисперсией = 1**. 

Чтобы привести шкалу к такому виду, из каждого наблюдения нужно отнять выборочное среднее значение и разделить на выборочное стандартное отклонение. 

$$ Z_{i}=\frac{x_{i} - \bar{x}}{std} $$

### 6.2 Для чего нужна Z-стандартизация?
В статистике чаще всего рассматривают именно нормальное распределение со средним = 0 и дисперсией = 1. 

Например, статистические тесты, предполагающие нормальность распределения, подразумевают проведение стандартизации для расчета статистики.

Иногда нам необходимо рассчитать z-значение только для отдельно взятого наблюдения, 
чтобы выяснить, **насколько далеко оно отклоняется от среднего значения в единицах стандартного отклонения**.

### 6.3 Правило 3-х сигм

<img src="images/6. 3 sigma rule.png" width="500px"/>

$$ M_{x} \pm \sigma \approx 68\text{% всех наблюдений} $$
$$ M_{x} \pm 2\sigma \approx 95\text{% всех наблюдений} $$
$$ M_{x} \pm 3\sigma \approx 100\text{% всех наблюдений} $$

В зависимости от значения стандартного отклонения график распределения может быть более узким или более широким, 
но распределение в любом случае является нормальным и подчиняется правилу трех сигм.


**Пример:**

```python
# Считается, что значение IQ (уровень интеллекта) у людей имеет нормальное распределение
# со средним значением равным 100 и стандартным отклонением равным 15 (M = 100, sd = 15).
# Какой приблизительно процент людей обладает IQ > 125?

from scipy import stats
mean = 100
std = 15
IQ=125

# sf - Survival function = (1 - cdf) - Cumulative distribution function
print(f'Только {(stats.norm(mean, std).sf(IQ))*100:.2f}% людей имеют IQ > {IQ}')
```

## **7. Центральная предельная теорема (ЦПТ)**

<img src="images/7. CPT mindmap.jpg" width="300px"/>

### 7.1 ЦПТ (Central Limit Theorem)

Допустим, у нас есть **какое-то** распределение со средним **M** и **известной конечной дисперсией**. 

ЦПТ гласит, что если мы будем многократно извлекать из генеральной совокупности выборки одинакового размера и считать у каждой из них выборочное среднее,
то **распределение этих выборочных средних будет нормальным, а среднее значение всех полученных выборочных средних будет равно среднему ГС = М**.

Стандартное отклонение полученного распределения выборочных средних называется **стандартной ошибкой среднего**, обозначается как **se** и буквально означает,
насколько сильно в среднем выборочные средние отклоняются от среднего ГС.

$$ se=\frac{\sigma}{\sqrt{n}} $$

*Если размер выборок, которые мы берем из ГС, **больше 30 (n > 30)**, то принято считать, 
что вместо стандартного отклонения ГС ($\sigma$) для расчета **оценки se** можно использовать выборочное стандартное отклонение **std**.

$$ se=\frac{std}{\sqrt{n}} $$

Чем больше размер выборок, которые мы берем из ГС, тем репрезентативнее каждая из выборок и тем меньше стандартная ошибка среднего **se**.
**То есть, чем больше размер выборок, тем большее количество полученных выборочных средних лежат очень близко к среднему ГС** 
(распределение выборочных средних станет **более узким**).

### 7.2 Чем может быть полезна ЦПТ?
Пусть у нас есть только **одна** выборка, полученная из ГС.

Имея размер этой выборки **n**, выборочное стандартное отклонение **std** и посчитав ее **выборочное среднее**, мы можем сказать, 
что **среднее ГС равно выборочному среднему со стандартной ошибкой среднего se**.

То есть, ЦПТ позволяет оценить среднее ГС, имея всего одну выборку.

**Пример:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# значения игральной кости
dice = [1, 2, 3, 4, 5, 6]
# количество бросков кости
count = 6
# размер генеральной совокупности
sp_size = 10000
# sp - Statistical population - генеральная совокупность
plt.figure(figsize=(5,3))
sp = pd.Series(dtype=np.int64, index=range(sp_size))
for i in range(sp_size):
    value = 0
    for _ in range(count):
        value += np.random.choice(dice)
    sp[i] = value

sp.plot.hist(bins=28);
```

```python
# количество выборок
samples_count = 10
# размер выборки
sample_size = 200
samples = pd.DataFrame([
    [np.random.choice(sp) for _ in range(sample_size)] for __ in range(samples_count)
]).T

samples.hist(figsize=(16, 10), sharex=0)
plt.subplots_adjust(hspace = 0.6);

# ЦПТ
means = samples.mean()
print(f'Population mean = {sp.mean()}, mean of sample means = {means.mean()}, se = {means.std()}')

# Возьмем произвольную выборку 
sample = samples[0]
print('Sample mean:', sample.mean())
print('Sample se: ', sample.std() / math.sqrt(sample.size))
```

## **8. Доверительные интервалы для среднего**

<img src="images/8. CI mindmap.jpg" width="300px"/>

### 8.1 В чем суть доверительного интервала?

ДИ позволяет проводить **интервальную оценку** какого-либо показателя.

Например, мы хотим оценить среднее ГС по выборке.

 - **Точечной** оценкой среднего ГС будет **выборочное среднее $\bar{x}$**
 - **Интервальной** оценкой среднего ГС будет **интервал [..., ...]**, в который среднее ГС будет попадать с заданной вероятностью

**Пример, отражающий суть доверительного интервала:**

Если мы возьмем **100 выборок из ГС** и построим для **каждого выборочного среднего 95% ДИ**, 
то в **95 случаях из 100 среднее ГС попадет в построенные интервалы**.

Другими словами, если мы строим 95% доверительный интервал для выборочного среднего, 
этот **ДИ вида $[\bar{x} - 1.96*se; \bar{x} + 1.96*se]$ будет включать в себя среднее ГС с 95% вероятностью**.

Еще другими словами, если мы строим 95% ДИ для выборочного среднего, **мы можем быть на 95% уверены, что среднее ГС принадлежит рассчитанному доверительному интервалу**.

### 8.2 Как построить доверительный интервал для выборочного среднего?
1. Основываясь на нашей выборке, считаем **стандартную ошибку среднего se**
2. Ищем левую и правую границы ДИ, основываясь на **желаемой ширине ДИ (чаще всего строят 90%, 95% или 99% ДИ)**
3. 95% ДИ для выборочного среднего $\bar{x}$ равен $[\bar{x} - 1.96*se; \bar{x} + 1.96*se]$ 

### 8.3 Откуда взялось число 1.96?
1.96 - это **критическое значение**, которое можно **запомнить** или найти в **таблице критических z-значений** 
(но можно посчитать самостоятельно с помощью интегрирования функции плотности вер-ти нормального распределения).

**Примеры с критическими значениями для других ДИ:**
 - **90%** всех наблюдений лежат в интервале $[\bar{x} - 1.65*se; \bar{x} + 1.65*se]$
 - **95%** всех наблюдений лежат в интервале $[\bar{x} - 1.96*se; \bar{x} + 1.96*se]$
 - **99%** всех наблюдений лежат в интервале $[\bar{x} - 2.58*se; \bar{x} + 2.58*se]$

*Чем **точнее** мы хотим оценить среднее ГС, тем более **узкий ДИ** нам нужно построить. 90% ДИ **уже**, чем 95% ДИ.

**Пример построения 99% ДИ:**

```python
# Рассчитайте 99% доверительный интервал для следующего примера: 
# среднее = 10, стандартное отклонение = 5, размер выборки = 100

from numpy import sqrt
from scipy import stats

p = 0.99
mean = 10
std = 5
n = 100

se = std/sqrt(n)
alpha = (1-p)/2
sigma = stats.norm().isf(alpha)
сonfidence_interval = mean - sigma * se, mean + sigma * se
print('[%.2f; %.2f]' % сonfidence_interval)
```

## **9. Идея статистического вывода, p-уровень значимости**

Чаще всего в статистике требуется не просто оценить какой-то параметр (например, среднее), а проверить определенную гипотезу.

### 9.1 Идея проверки статистических гипотез на примере
**Дано:**

На рынке есть лекарство от какой-то болезни, о котором известно, что оно позволяет вылечить человека в **среднем за 20 дней** ($M = 20$).

Мы разрабатываем свое лекарство, проверяем его на выборке из **64 человек** ($n = 64$) и получаем, 
что **средний срок выздоровления равен 18.5 дней** ($\bar{x} = 18.5$) со **стандартным отклонением, равным 4** ($std = 4$).

**Что хотим?**

Хотим понять, **позволяет ли действительно наше лекарство снизить срок выздоровления**.

Другими словами, у нас есть **выборка** и известная **константа** ($M = 20$), а мы хотим понять, 
**если мы дадим наше лекарство не 64 людям, а десяткам тысяч больных (обозначим как $M_{x}$, хотя обычно $M_{x}\equiv\bar{x}$), срок выздоровления будет меньше, чем заданное среднее (константа), или нет?**

**Решение:**

Для начала мы формулируем **нулевую гипотезу H0** и **альтернативную гипотезу H1**.

**H0:** между заданным средним и средним сроком выздоровления людей, воспользовавшихся нашим лекарством, **на самом деле нет разницы** ($M_{x}=20$).

**H1:** заданное среднее и среднее время выздоровления людей, воспользовавшихся нашим лекарством, **действительно не равны** ($M_{x} \neq 20$).

Теперь **предположим, что Н0 верна**, и воспользуемся **ЦПТ:**

Согласно ЦПТ, полученное **выборочное среднее** ($\bar{x}$) распределено вокруг **среднего ГС** ($M_{x}$) **с какой-то стандартной ошибкой среднего** $se$.

Посчитаем стандартную ошибку среднего по формуле:
$$ se=\frac{std}{\sqrt{n}}=\frac{4}{\sqrt{64}}=0.5 $$

Т.к. мы исходим из того, что **Н0 верна**, $M_{x} = 20$. 
Теперь посмотрим, насколько сильно наше выборочное среднее ($\bar{x}$) отклонилось от среднего ГС ($M_{x}$) в терминах стандартных отклонений.

Для этого проведем **стандартизацию** и посчитаем Z-значение по формуле:
$$ Z=\frac{\bar{x} - M_{x}}{se}=\frac{\bar{x} - M}{se}=\frac{18.5 - 20}{0.5}=-3 $$

Получается, что полученное выборочное среднее $\bar{x}$ отклонилось от среднего ГС $M_{x}$ на **3 стандартных ошибки среднего** $se$ влево.

Теперь, воспользуемся свойствами нормального распределения и посчитаем вероятность **такого или еще более сильного отклонения от среднего ГС** (в данном случае рассмотрим отклонение в обе стороны):

$ P(X < -3 \space or \space X > 3)=0.0027 $

Получается, что **если Н0 верна** ($ M_{x}=M $), то выборочные средние распределились бы вокруг заданного среднего ($M_{x}=M=20$) со стандартной ошибкой среднего $se=0.5$.

В нашем случае выборочное среднее $\bar{x}$ равно 18.5, что отличается от заданного среднего $M=M_{x}=20$ на 3 $se$. 

При этом **вероятность получить такое или еще более сильное отклонение случайно** равна $ 0.0027 \approx 0.003 $ и **называется p-value**.

### 9.2 Что делать с p-value?

<img src="images/9. p-value.jpeg" width="500px"/>

Сначала (еще до эксперимента) нам нужно задать уровень значимости $\alpha$ (он же называется вероятностью ошибки 1-го рода).

Как правило, задают $ \alpha=0.05 $, но если требуется более точный ответ, могут взять $ \alpha=0.01 $.

Если полученный **p-value оказался меньше $\alpha$**, то **можно отклонять нулевую гипотезу Н0 и принимать альтернативную Н1**.

Если же полученный **p-value оказался больше $\alpha$**, то у нас **недостаточно оснований для отклонения Н0**.

*Сама величина p-value ничего не говорит о вероятности того, что Н0 верна или не верна.

*p-value не дает нам **никаких гарантий**, потому что если мы будем повторять наш эксперимент очень много раз,
рано или поздно мы получим статистически значимые различия **абсолютно случайно**.

**Пример решения:**

```python
from scipy import stats

M = 20
mean = 18.5
n = 64
std = 4
alpha = 0.05

# Calculating standard error (se)
se = std / n ** 0.5

# Calculate Z
z = (mean - M) / se

# Calculate p-value (two-sided hypothesis)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z)))

if p_value > alpha:
    print(f'p-value = {p_value} > {alpha} => Can not reject H0')
else:
    print(f'p-value = {p_value} < {alpha} => Reject H0')
```

### 9.3 Кратко про ошибки 1-го и 2-го рода

**Вероятность ошибки 1-го рода ($\alpha$)** - это вероятность **отклонить Н0 и принять Н1**, когда на самом деле **верна Н0**.

Простыми словами, **ошибка 1-го рода** - это когда мы делаем **ложноположительный** вывод, сказав, что различия **есть**, когда на самом деле их **нет**.

**Вероятность ошибки 2-го рода ($\beta$)** - это вероятность **не отклонить Н0**, когда на самом деле **Н0 не верна**.

Простыми словами, **ошибка 2-го рода** - это когда мы делаем **ложноотрицательный** вывод, сказав, что различий **нет**, когда на самом деле они **есть**.

### 9.4 Проверка гипотез с помощью доверительного интервала
**Дано:**
$\bar{x}=18.5, \space std=4, \space n=64$

**Логика:** Строим 95% доверительный интервал (в случае, если берем $\alpha=0.05$) для $\bar{x}$.

Если заданное среднее ($M=20$) **попадает** в построенный доверительный интервал, то у нас **нет оснований отклонить Н0**.

И наоборот, если заданное среднее ($M=20$) **не попадает** в построенный доверительный интервал, то **можно отклонять Н0 и принимать Н1**.

```python
from scipy import stats

n = 64
std = 4
mean = 18.5
M = 20
p = 0.95
alpha = (1 - p) / 2

k = stats.norm.isf(alpha)
se = std / n ** 0.5
сonfidence_interval = mean - k * se, mean + k * se
print('[%.2f; %.2f]' % сonfidence_interval)

if (M >= mean - k * se) & (M <= mean + k * se):
    print(f'M is inside сonfidence_interval => Can not reject H0')
else:
    print(f'M is outside сonfidence_interval => Reject H0')
```

## **10. Т-распределение**

<img src="images/10. when to use t-distribution.jpg" width="300px"/>

Как уже обсуждалось ранее, ЦПТ прекрасно работает, **если размер выборки большой**, 
потому что в таком случае стандартное отклонение выборки точно оценивает стандартное отклонение ГС.

Но что, если у нас маленькая выборка?

### 10.1 Распределение Стьюдента (t-distribution)
Если в выборке **мало** элементов и $\sigma$ **неизвестна** (чаще всего), используется распределение Стьюдента.

**Оно:**
1. Унимодально
2. Симметрично
3. Имеет более **тяжелые хвосты**, т.е. наблюдения с **большей** вер-ю попадают за пределы $M \pm 2\sigma$

Форма распределения определяется **числом степеней свободы (df, degrees of freedom)**, которое равно $df=n-1$. 
Чем **больше** степеней свободы - тем ближе распределение к нормальному.

При **df > 30** распределение Стьюдента приближается к нормальному.


```python
# График снизу показывает, как меняется форма распределения при увеличении количества степеней свободы.

from scipy.stats import t, norm
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 100)
y1, y2, y3 = t.pdf(x, df=1), t.pdf(x, df=3), t.pdf(x, df=10)
y4 = norm.pdf(x)

plt.figure(figsize=(5,3))
plt.title('t-distribution with different df')
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4, 'r:')
plt.legend(('df=1', 'df=3', 'df=10', 'norm'))
plt.show()
```

### 10.2 Когда использовать распределение Стьюдента, а когда нормальное?
На самом деле, t-распределение стоит использовать не только в случае **маленькой выборки**, 
но и в случае, когда нам **неизвестна** $\sigma$ (стандартное отклонение ГС).

Нормальное распределение стоит использовать, если размер выборки **больше 30** и **известна** $\sigma$.

Соответственно, когда чуть выше мы рассчитывали $Z$, используя **выборочное** стандартное отклонение, а не $\sigma$, мы были **неправы**.

**Коректный способ расчета:**
$$ se=\frac{\sigma}{\sqrt{n}}, \space Z=\frac{\bar{x} - M} {se}=\frac{\bar{x} - M} {\frac{\sigma}{\sqrt{n}}}$$

$$ se=\frac{std}{\sqrt{n}}, \space t=\frac{\bar{x} - M} {se}=\frac{\bar{x} - M} {\frac{std}{\sqrt{n}}}$$

**Важный комментарий о ЦПТ:**

Если мы многократно выбираем из ГС выборки размера **менее 30**, 
то распределение полученных выборочных средних будет не **нормальным**, а **t-распределением** с df=n-1.

### 10.3 Чуть подробнее о степенях свободы или почему df=n-1
**Логика степеней свободы заключается в "единицах информации":**

Если у нас есть выборка из 10 элементов (n=10) и мы знаем среднее, 
то нам достаточно было бы знать только 9 элементов (df=10-1=9) и среднее, чтобы понять, чему равен 10 элемент.

**Как нам это помогает?**

Если мы посчитали t-значение и оно оказалось, например, равным **2** (отклонение на 2 станд. отклонения),
то в зависимости от количества степеней свободы мы можем получить абсолютно разную интерпретацию результата.

*Подробнее специфика степеней свободы будет рассмотрена во 2 части курса.


```python
# На выборке в 15 наблюдений при помощи одновыборочного t-теста проверяется нулевая гипотеза: μ=10 
# Если рассчитанное t-значение равняется -2 (t = -2), то p-уровень значимости  (двусторонний) равен:

from scipy import stats

t = -2
n = 15
df = n - 1

p = 2 * stats.t.sf(abs(t), df)
print(f'p = {p:.3f}')
```

## **11. Cравнение двух средних, t-критерий Стьюдента**

### 11.1 Зачем нужен t-тест (t-критерий)?
Он позволяет нам сравнивать два выборочных средних (то есть сравнивать средние двух выборок между собой).

### 11.2 T-критерий

**Задача:**

У нас есть две выборки:

- $ std_{1}, \space n_{1}, \space \bar{x_{1}} $
- $ std_{2}, \space n_{2}, \space \bar{x_{2}} $

Мы хотим сравнить **выборочные средние** $ \bar{x_{1}} $ и $ \bar{x_{2}} $, чтобы с их помощью сделать выводы о **средних ГС** $M_{1}$ и $M_{2}$.

**Сформулируем гипотезы:**

- **Н0:** $M_{1} = M_{2}$ (выборки принадлежат одной ГС)
- **Н1:** $M_{1} \neq M_{2}$ (выборки принадлежат разным ГС)

Предположим, что **Н0 верна**, и воспользуемся **ЦПТ**:

Если мы многократно будем брать по **паре выборок** и считать **разность их средних** ($ \bar{x_{1}}-\bar{x_{2}} $),
то ($ \bar{x_{1}}-\bar{x_{2}} $) распределится вокруг **разности средних ГС ($ M_{1}-M_{2} = 0 $)** как **t-распределение** со стандартной ошибкой среднего,
равной:

$$ se=\sqrt{\frac{std_{1}^2}{n_{1}} + \frac{std_{2}^2}{n_{2}}} $$

и **количеством степеней свободы df**, равным:

$$ df = n_{1} + n_{2} - 2 $$

При этом само **t-значение рассчитыватся так:**

$$ t=\frac{(\bar{x_{1}} - \bar{x_{2}}) - (M_{1} - M_{2})} {se} 
= \frac{(\bar{x_{1}} - \bar{x_{2}}) - (M_{1} - M_{2})} {\sqrt{\frac{std_{1}^2}{n_{1}} + \frac{std_{2}^2}{n_{2}}}}
= \frac{(\bar{x_{1}} - \bar{x_{2}}) - 0} {\sqrt{\frac{std_{1}^2}{n_{1}} + \frac{std_{2}^2}{n_{2}}}}
= \frac{(\bar{x_{1}} - \bar{x_{2}})} {\sqrt{\frac{std_{1}^2}{n_{1}} + \frac{std_{2}^2}{n_{2}}}}$$

Далее, основываясь на значениях $df$ и $t$ рассчитать **p-value**, 
показывающее **вероятность получить такое или еще более выраженное различие между двумя средними, если верна Н0**.

**Пример:**

```python
import pandas as pd
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np

array1 = np.array([84.7, 105.0, 98.9, 97.9, 108.7, 81.3, 99.4, 89.4, 93.0,
                   119.3, 99.2, 99.4, 97.1, 112.4, 99.8, 94.7, 114.0, 95.1, 115.5, 111.5])
array2 = np.array([57.2, 68.6, 104.4, 95.1, 89.9, 70.8, 83.5, 60.1, 75.7,
                   102.0, 69.0, 79.6, 68.9, 98.6, 76.0, 74.8, 56.0, 55.6, 69.4, 59.5])

# Считаем количество элементов, среднее, стандартное отклонение и стандартную ошибку
df = pd.DataFrame({'Sample 1': array1, 'Sample 2': array2}).agg(['mean','std','count','sem']).transpose()
df.columns = ['mean','std','n','se']

# Рассчитываем 95% интервал отклонения среднего
p = 0.95

K = t.ppf((1 + p)/2, df['mean']-1)
df['interval'] = K * df['se']

df
```

```python
# Cтроим 2 графика
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Boxplot
bplot1 = ax1.boxplot([array1, array2],
                     vert=True,  # Creating vertical boxes
                     patch_artist=True, # Coloring
                     labels=['Sample 1', 'Sample 2']) # Setting labels

# CI
bplot2 = ax2.errorbar(x=df.index, y=df['mean'], yerr=df['interval'],\
                      color="black", capsize=3, marker="s", markersize=4, mfc="red", mec="black")

# Coloring boxplot  
colors = ['pink', 'lightgreen']
for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)
    
# Adding info to both charts
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_title('Melting Temperature of 2 Types of DNA')
    # ax.set_xlabel('Comparing Samples')
    ax.set_ylabel('Temperature, F')
    
plt.show();
```

*Правый график, на котором отражены два выборочных средних и их доверительные интервалы, позволяет быстро проверить гипотезу.
Если среднее 1 выборки не попадает в пределы ДИ среднего 2 выборки и наоборот - **можно отклонять Н0**.

### 11.3 Пара уточнений по поводу t-теста
1. При проведении t-теста желательно, чтобы **соблюдалось условие гомогенности дисперсий**, т.е. чтобы дисперсии двух выборок были равны (или очень близки).
Для проверки этого условия можно использовать **критерий Левина**, **критерий Романовского** или **критерий Фишера**.

2. Если размер выборок небольшой (n<30), то важно, чтобы распределения обеих выборок были **нормальными**. 
Если n>30, то **нормальность распределений уже не так принципиальна**.

```python
# Рассчитайте доверительный интервал основываясь на знании t - распределения для среднего значения температуры плавления ДНК у первого вида:

from scipy import stats
from math import sqrt

mean = 89.9
std = 11.3
n = 20

# Degrees of freedom
df = n - 1

# 95% Confidence Interval
p = 0.95
alpha = 1 - p

# Standard Error
se = std / sqrt(n)

# ppf - Percent point function
t_critical = stats.t(df).ppf(1 - (alpha / 2))

# Confidence Interval 
сonfidence_interval = (mean - t_critical * se, mean + t_critical * se)
print('[%.2f; %.2f]' % сonfidence_interval)
```

```python
# Первые премии Оскар за лучшую мужскую и женскую роль были вручены в 1929. 
# Данные гистограммы демонстрируют распределение возраста победителей с 1929 по 2014 год (100 мужчин, 100 женщин). 
# Используя t-критерий проверьте, можно ли считать наблюдаемые различия в возрасте между лучшими актрисами и актерами статистически достоверными.
# Средний возраст мужчин равен 45, sd = 9.
# Средний возраст женщин равен 34, sd = 10.

# Sample 1
mean_1 = 45
std_1 = 9
n_1 = 100

# Sample 2
mean_2 = 34
std_2 = 10
n_2 = 100

p = 0.95
alpha = 1 - p

se = sqrt(std_1**2 / n_1 + std_2**2 / n_2)
df = n_1 + n_2 - 2
t = (mean_1 - mean_2) / se

# 1. Solving with 0.95 CI
t_critical = stats.t(df).ppf(1 - (alpha / 2))
ci_095 = [(mean_1 - mean_2) - t_critical * se, (mean_1 - mean_2) + t_critical * se]

if t > ci_095[1] or t < ci_095[0]:
    print(f't = {t:.3f} is not inside {ci_095} => Reject H0')
else:
    print(f"t = {t:.3f} is inside {ci_095} => Can't Reject H0")

# 2. Solving with t-test (not stats.ttest_ind because we don't have samples themselves, so we can just calculate p-value)
from scipy import stats

p_value = 2 * stats.t.sf(abs(t), df)

if p_value < alpha:
    print(f'p-value = {p_value:.3f} < {alpha:.2f} => Reject H0')
else:
    print(f"p-value = {p_value:.3f} > {alpha:.2f} => Can't Reject H0")
```

## **12. Проверка распределения на нормальность, QQ-Plot**

В статистике можно очень часто встретить требование **нормальности распределения**.

Какие есть способы проверить распределение на нормальность 
(другими словами, проверить, насколько наше эмпирическое распределение отличается от **теоретического нормального**)?

### 12.1 Гистограмма частот с наложением "идеальной" кривой
Это самый простой, но самый неточный вариант - мы строим обычную гистограмму частот, 
где по шкале Y отложены значения функции плотности вероятности, и **накладываем поверх гистограммы кривую "идеального нормального распределения"**.

После этого у нас получается визуально оценить разницу между нашим распределением и идеальным нормальным.

### 12.2 QQ Plot (Quantile-Quantile Plot)
Это способ более точный, чем первый, т.к. в нем банально больше информации.

По **оси Х** на нем отложена шкала нормального распределения **с нулем и отклонениями вправо и влево** ($ M \pm 3\sigma $).

По **оси Y** отложена обычная шкала нашего эмпирического распределения (не в терминах сигм).

Каждый элемент выборки представляет из себя **одну точку**, 
чем **ближе** точки к диагонали - тем **ближе распределение к нормальному**. 

Если наше эмпирическое распределение нормально - **все (или почти все) точки будут лежать на диагонали**.

**Пример интерпретации:**

 - Если точки отклоняются **вверх** от диагонали примерно на уровне 2 сигм по оси Х - значит в нашем распределении **слишком большие** 
 значения **в районе 2 сигм у идеального нормального распределения**.
 - Если точки отклоняются **вниз** от диагонали примерно на уровне 2 сигм по оси Х - значит в нашем распределении **слишком маленькие** 
 значения **в районе 2 сигм у идеального нормального распределения**.

<img src="images/12. qq plot examples.jpeg" width="900px"/>


```python
# Normality check using QQ plot and histogram

import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 4]

mu, sigma = 10, 4
n = 500 # с ростом числа точек в выборке qq-plot стремится к прямой
sequence = np.random.normal(mu, sigma, n)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('QQ Plot')

# Q-Q Plot graph
stats.probplot(sequence, dist="norm", plot=ax1)
ax1.set_title("Normal Q-Q Plot")

# normal distribution histogram + distribution
count, bins, _ = ax2.hist(sequence, 25, density=True)
p_x = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu) ** 2 / (2 * sigma ** 2) )
ax2.plot(bins, p_x, color='r')

plt.show();


# Shapiro-Wilk test
from scipy import stats
alpha = 0.05

p_value = stats.shapiro(sequence).pvalue

if p_value < alpha:
    print(f'p-value = {p_value:.3f} < {alpha:.2f} => Reject H0')
else:
    print(f"p-value = {p_value:.3f} > {alpha:.2f} => Can't Reject H0 (Sample Comes From a Normally Distributed Population)")
```

### 12.3 Тест Колмогорова-Смирнова и тест Шапиро-Уилка
Лучший способ проверки распределения на нормальность специальные статистические тесты.

Например, тест Шапиро-Уилка проверяет **Н0: выборка получена из нормального распределения**. Если мы получаем p-value > 0.05, то **Н0 верна**,
и наша выборка получена из **ГС, которая распределена нормально**.

### 12.4 Что такое выброс и как он может испротить нам жизнь?
**Выброс** - это **слишком большое** или **слишком маленькое** значение в выборке (значение, которое слишком "выбивается" из всех остальных).

На практике, если, например, мы проводим t-тест и хотим сравнить средние двух выборок, и в каждой из выборок есть всего **по одному** выбросу,
мы можем получить совершенно противоположные результаты (например, без этих двух выбросов p-value **был бы < 0.05**, а **с ними > 0.9**).

**С выбросами можно по-разному себя вести в зависимости от ситуации** - иногда, например, их можно удалить, но углубимся в эту тему позже.

### 12.5 U-тест Манна-Уитни (Mann-Whitney U-test)

<img src="images/12. mann-whitney vs t-test.png" width="500px"/>

Этот статистический тест служит тем же целям, что и t-тест - позволяет нам делать выводы о двух ГС, сравнивая две выборки.

**Тест Манна-Уитни:**
 - Не имеет требования нормальности распределений
 - Непараметрический (параметром в t-тесте является количество степеней свободы df)
 - Не столь чувствителен к выбросам по сравнению с t-тестом

Другими словами, если наши выборки получены из **совсем ненормальных распределений** или мы **имеем выбросы, 
которые не можем удалить** - можно использовать тест Манна-Уитни.

**Как он работает?**

Тест берет две наши выборки и **ранжирует элементы в них от самого маленького до самого большого**, 
после чего сравнивает **средний ранг первой группы со средним рангом второй**.

**Чем плох тест Манна-Уитни?**

1. Тест Манна-Уитни сравнивает **не средние двух выборок**, а **средние ранги выборок**, что достаточно тяжело **интерпретировать.**
2. Еще одна проблема теста Манна-Уитни - если мы проводим АБ тест и хотим воспользоваться формулой для расчета размера выборок/эффекта,
мы **не можем использовать формулу для t-теста**.
3. Тест Манна-Уитни даже в **идеальных условиях** (когда выборки имеют нормальное распределение) имеет **меньшую мощность** по сравнению с t-тестом.

*В целом, t-тест можно использовать и в случае, когда наши выборки получены не из нормального распределения, 
потому что его результаты значительно легче интерпретировать.


## **13. Однофакторный дисперсионный анализ (One-Way ANOVA)**

**t-критерий Стьюдента** и его непараметрические аналоги предназначен для сравнения исключительно **двух совокупностей**. 
Если их больше - мы можем применять однофакторный дисперсионный анализ. 
Та переменная, которая будет разделять наших испытуемых или наблюдения на группы
(**номинативная** переменная с нескольким градациями) называется **независимой переменной**. 
А та **количественная** переменная, по степени выраженности которой мы сравниваем группы, называется **зависимая переменная**. 

### 13.1 Общая идея дисперсионного анализа
У нас есть **m групп (выборок)** с **одинаковым** количеством элементов **n**. 
Мы хотим понять, **насколько группы различаются между собой (насколько отличаются средние ГС, из которых получены группы/выборки)**.

**Гипотезы (в случае, когда у нас 3 группы):**
 - **Н0:** $ M_{1} = M_{2} = M_{3} $ (группы/выборки получены из ГС с одинаковыми средними или из одной ГС)
 - **Н1:** $ M_{1} = M_{2} \neq M_{3} $ или $ M_{1} \neq M_{2} = M_{3} $ или $ M_{1} \neq M_{2} \neq M_{3} $ 

### 13.2 Почему "однофакторный"?
Потому что в этом виде анализа у нас есть только **один "фактор"**, по которому мы разделили группы между собой
(например, этим фактором может быть город - тогда первая группа - это люди из Москвы, вторая - люди из СПБ и т.д.)

### 13.3 Почему "дисперсионный"?
**Дисперсия** - это мера **изменчивости**, поэтому и в дисперсионном анализе мы смотрим на **изменчивость** данных.

Согласно идее дисперсионного анализа, **общая изменчивость данных** $ SST $ состоит из:
 - **Внутригрупповой изменчивости** ($SSW$, sum of squares within groups), 
 которая буквально означает, насколько сильно элементы выборки отличаются **между собой**.
 - **Межгрупповой изменчивости** ($SSB$, sum of squares between groups), 
 которая буквально означает, насколько сильно элементы выборки отличаются **от элементов остальных выборок**.
 
$$ SS_{total} = \sum_{j=1}^{p}{\sum_{i=1}^{n_j}{(x_{ij} - \bar{x})^2}} = SS_{between} + SS_{within} $$
$$ SS_{between} = \sum_{j=1}^{p}{n_j{(\bar{x}_j - \bar{x})^2}} $$
$$ SS_{within} = \sum_{j=1}^{p}{\sum_{i=1}^{n_j}{(x_{ij} - \bar{x}_j)^2}} $$

После расчета изменчивостей мы смотрим, насколько сильно общая изменчивость обеспечивается SSB или SSW
(если, например, $ SST=30 $, $ SSB=3 $, a $ SSW=27 $, то мы видим, что общая изменчивость объясняется изменчивостью **внутри групп**, а не **между группами**,
откуда можно сделать вывод о том, что группы **похожи** и наоборот).

### 13.4 Как сделать точный вывод?

Для этого надо **рассчитать F-значение** по формуле:

$$\large F=\frac{MS_{bg}}{MS_{wg}}=
\frac{\frac{SSB}{df_{SSB}}}{\frac{SSW}{df_{SSW}}}
=\frac{\frac{SSB}{m-1}}{\frac{SSW}{N-m}}
=\frac{\frac{SSB}{m-1}}{\frac{SSW}{n*m-m}} $$

$$ df_{SSB}=m-1 $$
$$ df_{SSW}=N-m=m*n - m $$

$ MS_{bg} $ - межгрупповой средний квадрат (intergroup mean square)

$ MS_{wg} $ - внутригрупповой средний квадрат (intragroup mean square)

Полученное F-значение имеет **распределение Фишера**. 
Имея F, мы можем рассчитать p-value (распределение Фишера **асимметрично**, и большая часть значений располагается **слева**, 
а отрицательных значений нет вообще, поэтому p-value - это $ P(X > F) $).

Дальше логика стандартная - сравниваем p-value с заданным нами уровнем значимости $\alpha$ и делаем вывод о верности **Н0**.

### 13.5 Важное уточнение о формулировании вывода
При отклонении Н0 единственное, что мы можем сказать - это что **как минимум две группы отличаются между собой**.

Для того, чтобы сформулировать более понятный вывод с пониманием, 
какие конкретно группы и как конкретно различаются - нужно рассмотреть **множественные сравнения в ANOVA** (следующая тема).

```python
from scipy import stats
import pandas as pd

# Samples (here each column is one sample)
data = pd.DataFrame({
           'a': [3, 1, 2],
           'b': [5, 3, 4],
           'c': [7, 6, 5]
          })

plt.figure(figsize=(5,3))
data.boxplot()
print('H0:', '='.join(data))
print('H1:', f'!({"=".join(data)})')
alpha = 0.05

# Total mean
grand_mean = data.values.flatten().mean()

# Deviation of group means from the total mean (SSB)
ssb = sum(data[group].size * (group_mean - grand_mean)**2  for group, group_mean in data.mean().items())

# Deviation of group items from the group means (SSW)
ssw = sum(sum((x - group_mean)**2 for x in data[group]) for group, group_mean in data.mean().items())

# Degrees of freedom
groups = data.shape[1] # m
dfb = groups - 1 # m-1
dfw = data.size - groups # N-m

# Intergroup square mean
mssb = ssb / dfb

# Intragroup square mean
mssw = ssw / dfw

# F-value
f_value = mssb / mssw

# Calculating p-value
p_value = stats.f.sf(f_value, dfb, dfw)

if p_value < alpha:
    print(f'p-value = {p_value:.3f} < {alpha:.2f} => Reject H0')
else:
    print(f"p-value = {p_value:.3f} > {alpha:.2f} => Can't Reject H0")
```

## **14. Множественные сравнения в ANOVA**

### 14.1 В чем вообще проблема?
Когда у нас не **две**, а **несколько** групп, для их сравнения мы можем использовать однофакторный дисперсионный анализ (One-Way ANOVA).
При этом, если мы **отклоняем Н0**, то мы не можем сделать вывод о том, **какая конкретно группа как отличается от другой**, 
и наш вывод звучит как **"как минимум две группы значимо отличаются"**, а мы хотим более конкретный вывод.

Для того, чтобы получить более конкретный вывод, можно использовать, например, **t-тест**,
и в случае, когда у нас 3 группы, мы делаем **несколько попарных сравнений** 
(в данном случае - 1 с 2, 1 с 3 и 2 с 3, т.е. 3 попарных сравнения).

По сути, каждое **попарное сравнение - это проверка новой гипотезы**:
 - Если сравниваем 1 и 2 группы, то **Н0:** $ M_{1} = M_{2} $
 - Если сравниваем 1 и 3 группы, то **Н0:** $ M_{1} = M_{3} $
 - Если сравниваем 2 и 3 группы, то **Н0:** $ M_{2} = M_{3} $

### 14.2 В чем проблема попарных сравнений?
Это легко показать на примере:

Пусть у нас есть ГС, из которой мы берем выборки и сравниваем их (т.е. мы заранее знаем, что **Н0 верна**).
- Когда мы сравниваем t-тестом **две выборки** и устанавливаем $ \alpha = 0.05 $, то мы совершаем ошибку 1-го рода в **~5% случаев**
(т.е. мы отклоняем Н0, когда она на самом деле верна, т.к. сравниваемые выборки действительно из одной ГС)
- Когда мы сравниваем t-тестом **больше двух выборок** и устанавливаем $ \alpha = 0.05 $, 
то мы перестаем контролировать вероятность ошибки 1-го рода на уровне **~5%**, и она **значительно возрастает** 
(т.к. если у нас, например, 10 попарных сравнений, то мы "прогоняем" тест 10 раз, 
тем самым **в 10 раз** увеличиваем вероятность **хоть в каком-то из "прогонов" случайно получить значимые различия, если их нет**)

Чтобы избавиться от этой проблемы, нужно использовать **поправки на множественные сравнения**.

### 14.3 Поправка Бонферрони

**Суть работы поправки**

Мы считаем **количество попарных сравнений** (кол-во "прогонов" стат. теста) и **делим установленный нами уровень значимости $\alpha$ на это число**.

Кол-во попарных сравнений вычисляет по комбинаторной формуле расчета:

$$C_n^k = \frac{n!}{(n-k)!\cdot k!}$$
где $n$ - всего элементов в выборке, а $k$ - количество выбираемых. 

**В чем проблема поправки?**

Хоть поправка и очень проста и позволяет **снизить вероятность ошибки 1 рода**, она в то же время значительно **увеличивает вероятность ошибки 2 рода**
(т.е. верноятность сделать **ложноотрицательный** вывод / **НЕ отклонить Н0**, когда на самом деле ее **нужно было отклонить**).

Из-за этого получается, 
что после применения поправки Бонферрони **практически невозможно отклонить Н0 и получить статистически значимые различия между группами**.

Для преодоления проблем, связанных с низкой мощностью метода Бонферрони, 
в 1978 г. Стур Холм (Holm) предложил гораздо более мощную его модификацию 
(часто этот метод называют еще **[методом Холма-Бонферрони](https://en.wikipedia.org/wiki/Holm%E2%80%93Bonferroni_method)**).

```python
from numpy import random

def pair_t(samples, alpha):
    '''Парный t-критерий. Если средние выборок равны, возвращает True'''
    n_samples = samples.shape[0]
    # https://ru.wikipedia.org/wiki/Сочетание    
    n_combinations = n_samples*(n_samples - 1)//2
    result = np.zeros(n_combinations, dtype=bool)
    k = 0
    for i in range(n_samples):
        for j in range(i+1, n_samples):
            N = samples[i].size
            std_err = np.sqrt((samples[i].std()**2)/N + (samples[j].std()**2)/N)
            t_value = (samples[i].mean() + samples[j].mean())/std_err
            p = stats.t.sf(t_value, N-2)
            result[k] = p >= alpha
            k += 1
    return np.all(result)

def pair_t_test(repeat, n_samples, sample_size, ax, alpha=0.05):
    '''
    Функция показывает, сколько у нас будет ложныхположительных результатов при попарном сравнении выборок с помощью t-критерия
    
    repeat, n_samples, sample_size = количество повторов, количество выборок в каждом повторе, размер выборки
    
    ax - для рисования
    alpha = уровень значимости
    '''
    result = np.zeros(repeat, dtype=bool)
    for i in range(repeat):
        samples = random.randn(n_samples, sample_size)
        result[i] = pair_t(samples, alpha)
    
    unique, counts = np.unique(result, return_counts=True)
    percentage = counts/result.size
    ax.pie(percentage, normalize=False, labels=unique, autopct='%.0f%%')

fig, axs = plt.subplots(ncols=4, figsize=(20, 4))
n_samples = [2, 4, 8, 16]
fig.suptitle('Type 1 Error With Paired T-test')

for n, ax in zip(n_samples, axs):
    pair_t_test(1000, n, 100, ax)
    ax.set_title(f'{n} samples')
```

Как можно заметить, при сравнении **двух** выборок с помощью t-теста с $\alpha=0.05$ вероятность ошибки 1 рода равна **~5%**.
Если мы возмём **4** выборки и сравним их попарно, то ошибка возрастёт в **4** раза до **20%**. 
При **8** выборках вероятность ошибки 1 рода возрасла почти в **9** раз - до **48%**.
**16** выборок дают увеличение вероятности до **81%** (почти в **16** раз).

```python
fig, axs = plt.subplots(ncols=4, figsize=(20, 4))
n_samples = [2, 4, 8, 16]
fig.suptitle('Type 1 Error With Paired T-test With Bonferroni Correction')

for n, ax in zip(n_samples, axs):
    alpha = 0.05/((n*(n-1))/2)
    pair_t_test(1000, n, 100, ax, alpha)
    ax.set_title(f'{n} samples')
```

### 14.4 Критерий Тьюки (Tukey HSD)

Критерий Тьюки стоит использовать **вместе с ANOVA**, чтобы в случае отклонения Н0 ($p < \alpha$) мы могли оценить,
**как именно различаются группы между собой**.

Сам критерий Тьюки очень похож на обычный t-тест, в котором мы попарно сравниваем выборки.
Разница между ними в том, что в критерии Тьюки по-другому рассчитывается стандартная ошибка.

Критерий Тьюки позволяет нам рассчитать **доверительные интервалы для разности каждой пары выборочных средних**, 
и если в доверительный интервал **не попадает 0** - отклоняем **Н0**.

**Вычисление критерия Тьюки:**

$\displaystyle q = \frac{\overline x_B - \overline x_A}{se}$, при этом $\displaystyle se = \sqrt \frac{MS_w}{n}$, где $\displaystyle MS_w$ - внутригрупповая дисперсия.

Приведённая формула верна для случаев, когда все сравниваемые группы содержат одинаковое число наблюдений $n$. 
Если же сравниваемые группы неодинаковы по размеру, 
то согласно [этому](https://www.itl.nist.gov/div898/handbook/prc/section4/prc471.htm), 
расчёт производится для каждой пары групп независимо (т.е. для групп $i$,$j$):

$$se_{ij} = \displaystyle\sqrt{\frac{MS_{{w}_{ij}}}{2}\left (\frac{1}{n_i}+\frac{1}{n_j}\right )}$$

Для расчёта доверительного интервала используется **не t-распределение, а "Studentized range distribution" со степенями свободы 2 и $n_i + n_j - 2$**.
Например, 95% довертельный интервал:  
$$\overline x_i - \overline x_j \pm q_{(0.05;2;n_i + n_j - 2)}{se}_{ij}$$

```python
from statsmodels.stats.multicomp import MultiComparison

data = pd.read_csv('genetherapy.csv')

MultiComp = MultiComparison(data['expr'], data['Therapy'])

print(MultiComp.tukeyhsd().summary())
```

## **15. Многофакторный ANOVA**

Многофакторный дисперсионный анализ отличается от однофакторного только **количеством факторов**, 
т.е. количеством **номинативных переменных**, по которым разделены наши выборки (например, пол, возраст, город проживания и т.д.).

### 15.1 Двухфакторный ANOVA
Помимо того, что факторов **два**, а не **один**, отличается от однофакторного анализа и метод расчета **общей изменчивости** $SST$:
$$ SS_{total} = SSW + SSB_{A} + SSB_{B} + SSB_{A} \cdot SSB_{B} $$

**Таким образом, общая изменчивость $SST$ обусловлена:**
- Внутригрупповой изменчивостью $SSW$
- Влиянием первого фактора $SSB_{A}$
- Влиянием второго фактора $SSB_{B}$
- **Взаимодействием этих двух факторов** $SSB_{A} \cdot SSB_{B}$ 
(влияние одного фактора по-разному проявляется на зависимую переменную в зависимости от градаций другого фактора)

Причиной **отклонения Н0 ($ M_{1} = M{2} = ... = M_{i} $)** в многофакторном дисперсионном анализе (о равенстве средних во всех группах) может стать:
- Значимый эффект одного из фактора
- Значимое взаимодействие факторов
- Значимый эффект для всех факторов

*Заметные выбросы могут значительно повлиять на результаты дисперсионного анализа, 
т.к. могут повлиять на значения средних межгрупповых и внутригрупповых квадратов (особенно в случае небольшого количества наблюдений).

**Пример решения задачи без библиотек:**


```python
import pandas as pd
from statsmodels.graphics.factorplots import interaction_plot
from scipy import stats

# Connecting to data
data = pd.read_csv('atherosclerosis.csv', sep=',')

# Plot
fig, ax = plt.subplots(figsize=(5, 4))
fig = interaction_plot(data.dose, data.age, data.expr, colors=['green','red'], markers=['D','^'], ms=10, ax=ax)

# Calculating degrees of freedom
N =  len(data.expr)
m1 = len(data.age.unique())
m2 = len(data.dose.unique())
df_a = m1 - 1
df_b = m2 - 1
df_axb = df_a * df_b 
df_w = N - m1 * m2

# Grand mean
grand_mean = data['expr'].mean()

# Sums of squares
ssq_a = sum([(data[data.age == i].expr.mean() - grand_mean) ** 2 for i in data.age])
ssq_b = sum([(data[data.dose == i].expr.mean() - grand_mean) ** 2 for i in data.dose])
ssq_t = sum((data.expr - grand_mean) ** 2)
spl_age=[data[data.age == i] for i in data.age.unique()]
age_means=[[x_age[x_age.dose == d].expr.mean() for d in x_age.dose] for x_age in spl_age]
ssq_w = sum([sum((spl_age[i].expr-age_means[i]) ** 2) for i in range(len(data.age.unique()))])
ssq_axb = ssq_t - ssq_a - ssq_b - ssq_w

# Mean squares
ms_a = ssq_a / df_a
ms_b = ssq_b / df_b
ms_axb = ssq_axb / df_axb
ms_w = ssq_w / df_w

# F-values
f_a = ms_a / ms_w
f_b = ms_b / ms_w
f_axb = ms_axb / ms_w

# P-values
p_a = stats.f.sf(f_a, df_a, df_w)
p_b = stats.f.sf(f_b, df_b, df_w)
p_axb = stats.f.sf(f_axb, df_axb, df_w)

# Results table
results = {'sum_sq':[ssq_a, ssq_b, ssq_axb, ssq_w],
           'df':[df_a, df_b, df_axb, df_w],
           'F':[f_a, f_b, f_axb, 'NaN'],
            'PR(>F)':[p_a, p_b, p_axb, 'NaN']}
columns=['sum_sq', 'df', 'F', 'PR(>F)']
aov_table1 = pd.DataFrame(results, columns=columns,
                          index=['age', 'dose', 
                          'age:dose', 'Residual'])
print(aov_table1)
```

### 15.2 Условия использования дисперсионного анализа

**ANOVA требует выполнения условий:**
- Нормальность распределения зависимой переменной в каждой из групп;
- Гомогенность дисперсий

**Гомогенность (однородность) дисперсий** - требование, чтобы дисперсия была примерно одиннаковой в каждой из групп.

Дисперсионный анализ еще и тем хорош, что при достаточно большом количестве наблюдений (**n>50**) ANOVA устойчива к нарушению **обоих** этих требований.

**Для проверки гомогенности дисперсий можно:**
1. Сначала построить box-plot и визуально оценить изменчивость переменной и наличие заметных выбросов.  
2. Применить [**тест Левина**](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.levene.html#scipy.stats.levene) 
и проверить нулевую гипотезу $H_0: D_1=D_2=D_n$ при p>0.05.  

**Для проверки распределений на нормальность можно:**
1. Сначала построить QQ-plot для каждой группы и визуально оценить отклонение наших выборок от нормального.  
2. Применить тесты **Шапиро-Уилка** или **Колмогорова-Смирнова**.

## **17. Понятие корреляции**

Корреляция позволяет нам исследовать **силу и направление взаимосвязи двух количественных переменных**, например:
- Рост и вес человека
- Возраст и количество тренировок в неделю
- Площадь квартиры и ее стоимость

Для расчета корреляции используется понятие **ковариация**.

### 17.1 Ковариация

(Ко - совместная, вариация - изменчивость). Мера **линейной** зависимости двух случайных величин.

Если **ковариация положительна**, то с ростом значений одной случайной величины, 
значения второй имеют тенденцию **возрастать**, а если знак **отрицательный — то убывать**.

$$ cov(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{x})}}{N - 1} $$
где N - количество случайных величин, а единица - количество степеней свободы.

Однако только по **абсолютному** значению ковариации **нельзя судить** о том, **насколько сильно величины взаимосвязаны**, 
так как масштаб ковариации зависит от их дисперсий. 
Значение ковариации можно нормировать, поделив её на произведение среднеквадратических отклонений (квадратных корней из дисперсий) случайных величин. 
Полученная величина называется **коэффициентом корреляции Пирсона**, который всегда находится в интервале [−1,1]:

$$ r(X, Y) = \frac{cov(X, Y)}{\sigma_x\sigma_y}$$

Для визуализации корреляции используется **диаграмма рассеивания (scatterplot)**, 
где каждая ось - это шкала значений каждой из выборок.

### 17.2 Вывод формулы коэффициента корреляции Пирсона

**Формула коэффициента корреляции Пирсона:**

$$ r(X, Y) = \frac{cov(X, Y)}{\sigma_x\sigma_y}$$

**Запишем формулу чуть подробнее и выполним возможные преобразования:**

$$ r(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{(N - 1)\sqrt{\sum{\frac{(x_i - \bar{x})^2}{N-1}}}\sqrt{\sum{\frac{(y_i - \bar{y})^2}{N-1}}}} $$

теперь вынесем 1/(N-1) из под корней 

$$ r(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{(N - 1)\frac{1}{(N-1)}\sqrt{\sum{(x_i - \bar{x})^2}}\sqrt{\sum{(y_i - \bar{y})^2}}} $$

и сократим (N-1)

$$ r(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2}}\sqrt{\sum{(y_i - \bar{y})^2}}} $$

таким образом, мы сократили (N-1) в знаменателе и получили финальную формулу для коэффициента корреляции, которую часто можно встретить в учебниках:

$$ r(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2}\sum{(y_i - \bar{y})^2}}} $$

### 17.3 Коэффициент детерминации

**Коэффициент детерминации** $R^2$:
- Показывает, в какой степени дисперсия одной переменной "обусловлена" влиянием другой переменной (взаимосвязью с другой переменной) 
- Равен **квадрату коэффициента корреляции Пирсона ($r_{x,y}^2$)**
- Принимает значения в интервале [0,1]

### 17.4 Проверка гипотез с помощью корреляции

**Идея:**

Имеем две выборки, хотим понять, есть ли корреляция **между ГС, из которых взяты эти выборки**.

**Гипотезы:**
- **Н0:** $r_{x,y} = 0$
- **Н1:** $r_{x,y} \neq 0$

Рассчитываем **t-значение** с количеством степеней свободы $ df=n-2 $, где $n$ - это количество элементов в одной выборке.

Для того чтобы проверить гипотезу, мы должны знать **распределение величины** $r_{x,y}$, 
но собственное распределение величины $r_{x,y}$ довольно сложное, поэтому мы применим преобразование:

$$ t=\frac{r_{x,y}}{\sqrt{1-r_{x,y}^2}}\sqrt{n-2} \approx t_{n-2}$$

Итак, **выборочное распределение этой статистики есть распределение Стьюдента с n-2 степенями свободы**
(подробнее в [этой статье](http://statistica.ru/theory/znachimost-koeffitsienta-korrelyatsii-doveritelnyy-interval/)).

**Проверка гипотезы и вывод:**

Т.к. наша статистика имеет распределение Стьюдента:
- Либо считаем критическое t-значение для заранее заданного нами уровнем значимости $\alpha$ и сравниваем полученное значение $t$ c $t_{critical}$
- Либо считаем p-value

**Пример:**

```python
# Covariation and correlation example

import numpy as np
import random as r

def cov(x, y):
    assert x.size == y.size
    return ((x - x.mean()) * (y - y.mean())).sum() / (x.size - 1)

def cor(x, y):
    return cov(x, y) / (np.std(x, ddof=1) * np.std(y, ddof=1))

def t_and_p_value(x, y):
    # Calculating t statistic for correlation and finding the p-value, ddof=n-2
    return 2 * stats.t.sf(abs(cor(x, y) / sqrt(1 - cor(x, y)**2) * sqrt(n - 2)), x.size-2) 

# Function which imitates random factor
# k - настолько существенным будет случайный фактор
def randomize(arr, k):
    alpha = np.max(arr) - np.min(arr)
    res = np.zeros(arr.shape)
    for i, v in enumerate(arr):
        sign = 1 if r.choice([True, False]) else -1
        res[i] = v + sign * alpha * r.random() * k
    return res

x = np.array(range(30))
y1 = randomize(x, 0.1)
y2 = randomize(x, 0.5)
y3 = randomize(x, 1)

import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 3))
ax1.scatter(x, y1)
ax2.scatter(x, y2)
ax3.scatter(x, y3)
ax1.set_title('High Correlation')
ax2.set_title('Mid Correlation')
ax3.set_title('Low Correlation')
plt.show();

print(f'''
cov1: {cov(x, y1):.2f}
cov2: {cov(x, y2):.2f}
cov3: {cov(x, y3):.2f}

cor1: {cor(x, y1):.2f}
cor2: {cor(x, y2):.2f}
cor3: {cor(x, y3):.2f}
''')

# Calculating p-value
from scipy import stats
alpha = 0.05

for idx, y in enumerate([y1, y2, y3]):
    if t_and_p_value(x, y) < alpha:
        print(f'p-value{idx+1} = {t_and_p_value(x, y):.3f} < {alpha:.2f} => Reject H0')
    else:
        print(f"p-value{idx+1} = {t_and_p_value(x, y):.3f} > {alpha:.2f} => Can't Reject H0")
```

## **19. Регрессия с одной независимой переменной**

В этом и следующих уроках мы научимся работать с **одномерным регрессионным анализом**, 
который позволяет проверять гипотезы о взаимосвязи **одной количественной зависимой переменной и одной/нескольких независимых**.

Сначала мы познакомимся с самым простым вариантом -  **простой линейной регрессией**, при помощи которой можно исследовать взаимосвязь двух переменных. 
Затем перейдем к **множественной регрессии** с несколькими независимыми переменными.

### 19.1 Регрессионный анализ

**Регрессионный анализ** - набор статистических методов исследования влияния **одной или нескольких независимых переменных** $X_{1}, X_{2} , ... , X_{p}$ 
на **зависимую переменную** $Y$.

Независимые переменные называют **регрессорами** или **предикторами**, 
а зависимые переменные — **критериальными** (эти названия **не отражают** причинно-следственные отношения) .

### 19.2 Линейная регрессия
Наиболее распространенный вид регрессионного анализа — **линейная регрессия (linear regression)**,
когда находят **линейную** функцию (прямую), которая, согласно определённым математическим критериям, наилучшим образом описывает имеющиеся данные 
(если, как в случае визуализации корреляции, построить scatterplot, то регрессионная прямая будет максимально "удачно" проходить через "облако точек").  

Лучшая аналогия линейной регрессии - это уравнение прямой из школьного курса: $ y = kx + b $.

**Формула линейной регрессии:**

$$\large \hat{y} = b_0 + b_1x $$
- $b_0$ **(intercept)** - значение пересечения линии с осью Y  
- $b_1$ **(slope)** задаёт угол наклона линии регрессии

Для расчета $b_0$ и $b_1$ часто применяется **метод наименьших квадратов**.

### 19.3 Метод наименьших квадратов

**Метод наименьших квадратов (МНК, Least Squares, LS)** - это способ нахождения оптимальных параметров линейной регресссии ($b_0$,  $b_1$), 
таких, что **сумма квадратов ошибок (остатков) была минимальна**. 

Расчёт параметров выполняется по формулам:
$$\large b_0 = \bar{y} - b_1\bar{x}, \space \space b_1 = \frac{std_y}{std_x}r_{x,y}$$
где $std$ - стандартное отклонение, а $r$ - коэффициент корреляции.

Сущность МНК (обычного, классического) заключается в том, чтобы найти такие параметры $b$, 
при которых **сумма квадратов отклонений** (**ошибок**, для регрессионных моделей их часто называют **остатками регрессии**) $e_{t}$ будет минимальной.

Другими словами, если мы построим диаграмму рассеивания (scatterplot), то **остатками** $e_{t}$ будут **расстояния от каждой точки до прямой**, 
которую мы с помощью параметров $b_0$ и $b_1$ пытаемся расположить так, чтобы эти расстояния были **наименьшими**. 

**Пример МНК:**


```python
# Least Squares Method

n = 100

X = np.array([np.random.randint(50,100) for i in range(n)])
Y = np.array([i + np.random.randint(80,120) - np.random.randint(-20,20) for i in X])

# Plotting
plt.figure(figsize=(6,3))
plt.scatter(X, Y)
b1, b0 = np.polyfit(X, Y, 1) #  b0 - intercept, b1 - slope
plt.plot(X, b0 + b1*X, color='red')

print("b0={}\nb1={}".format(b0,b1))
```

```python
# Calculate b_1
R = 0.25

# Samples
mean_x = 15
var_x = 25

mean_y = 10
var_y = 36

# Calculating r and std's
std_x = var_x ** 0.5
std_y = var_y ** 0.5
r = R ** 0.5

# Finding b_1 and b_0
b_1 = std_y / std_x * r
b_0 = mean_y + b_1 * mean_x

print(f'b_1 = {b_1}, b_0 = {b_0}')
```

## **20. Гипотеза о значимости взаимосвязи и коэффициент детерминации**

Как строить регрессионную прямую понятно, но как ответить на вопрос: "насколько статистически значима взаимосвязь двух наших переменных?"

Более подробно можно почитать [тут](http://statistica.ru/theory/osnovy-lineynoy-regressii/#%D0%9F%D1%80%D0%B5%D0%B4%D0%BF%D0%BE%D0%BB%D0%BE%D0%B6%D0%B5%D0%BD%D0%B8%D1%8F%20%D0%BB%D0%B8%D0%BD%D0%B5%D0%B9%D0%BD%D0%BE%D0%B9%20%D1%80%D0%B5%D0%B3%D1%80%D0%B5%D1%81%D1%81%D0%B8%D0%B8).

**Формулируем гипотезы:**
- $H_0:B_1=0$ (коэффициент $b_1$ в генеральной совокупности равен 0 => нет статистически значимой взаимосвязи между переменными)
- $H_1:B_1 \neq 0$ (коэффициент $b_1$ в генеральной совокупности НЕ равен 0 => есть статистически значимая взаимосвязь между переменными)

**Почему нас интересует именно коэффициент $b_1$?**

Взглянем на формулу расчета $b_1$:
$$ b_1 = \frac{std_y}{std_x}r_{x,y} $$

- Именно этот коэффициент определяет **угол наклона регрессионной прямой (slope)**, 
при этом именно **slope** позволяет наилучшим образом "описать" взаимосвязь переменных. 
- Если $b_1=0$, то и $b_0=0$, т.е. регрессионная прямая совпадает с горизонтальной прямой, проведенной через **среднее зависимой переменной Y** ($\bar{y}$),
при этом, согласно формуле для $b_1$, $b_1=0$ говорит о том, что корреляция между переменными $r_{x,y}=0$ - а это говорит об **отсутствии взаимосвязи между переменными**.

Далее, исходя из предположения о **верности Н0** и применения **ЦПТ**, считаем t-статистику:

$$t=\frac{b_1-B_1}{se}=\frac{b_1-0}{se}=\frac{b_1}{se}, \space \space df=N-2$$

**После расчета t-статистики можем:**
- Посчитать p-value и сравнить его с заранее заданным $\alpha$
- Найти критические значения t-распределения и проверить, попадает ли $B_1=0$ в ДИ с заданным уровнем значимости

### 20.1 Коэффициент детерминации
Это тот же самый коэффициент, который мы рассматривали в случае с **корреляцией Пирсона**.

$R^2 = r_{x,y}^2$ - описывает, какая часть дисперсии зависимой переменной Y "объясняется" нашей регрессионной моделью/влиянием независимой переменной.

**Наглядный пример для понимания $R^2$**

Пусть у нас есть 2 переменные:
- Качество (Y, зависимая переменная)
- Цена (Х, предиктор)

Мы хотим понять, как качество определяется ценой товара, 
т.е. **в какой мере изменчивость переменной "качество" объясняется взаимосвязью с переменной "цена"**?

Для начала, обратим внимание на **формулу $R^2$:**

$$R^2=1-\frac{SS_{res}}{SS_{total}}$$

- $SS_{res}$ - сумма квадратов расстояния от наблюдения до **регрессионной прямой** (на рисунке ниже - синие).
- $SS_{total}$ - сумма квадратов расстояния от наблюдения до **выборочного среднего $\bar{y}$** (на рисунке ниже - красные).

<img src="images/20. R^2 example.png" width="700px"/>

**Левый график:**

- **Синим** обозначена регрессионная прямая, которая **совсем чуть-чуть наклонена вверх**, 
потому что **коэффициент корреляции $r_{x,y} = 0.04$ слишком мал $\Rightarrow R^2 \rightarrow 0$**. 
- Наблюдения очень **далеко** расположены от регрессионной прямой $\Rightarrow$ сумма квадратов остатков $SS_{res}$ 
будет **почти равна** общей сумме квадратов $SS_{total}$.

**Правый график:** 
- Видим **ярко выраженную корреляцию ($r_{x,y}=0.99 \rightarrow 1 \Rightarrow R^2 \rightarrow 1$)**. 
- Наблюдения очень **близко** расположены к регрессионной прямой $\Rightarrow$ сумма квадратов остатков $SS_{res}$ 
будет **гораздо меньше** общей суммы квадратов $SS_{total}$.

Получается, что **чем лучше наша регрессионная модель** 
(чем лучше наша прямая описывает элементы выборок) - тем сильнее $\frac{SS_{res}}{SS_{total}} \rightarrow 0$ и тем сильнее $R^2 \rightarrow 1$, 
потому что $R^2 = 1 - \frac{SS_{res}}{SS_{total}}$.

$R^2 \approx 1$ говорит о том, что практически **100% изменчивости зависимой** переменной $Y$ 
обуславливается ее взаимосвязью с **независимой** переменной $X$.

### 21.1 Разница между корреляционной прямой и регрессионной прямой

- **Корреляционная прямая** показывает **направление** (положительное или отрицательное) и **"силу"** взаимосвязи, но в полной мере она входные данные **не описывает**.   
- **Регрессионная прямая** описывает наши данные, поэтому на основе этой прямой мы можем **предсказывать и те данные, которых не было во входных**.

**Например**:

У нас есть данные о цене и площади квартир в каком-то районе. 
- Корреляционная прямая может **показать взаимосвязь** площади квартиры с её ценой и мы, например, можем узнать, что чем больше площадь - тем выше цена. 
- Регрессионная прямая показывает **как площадь квартиры влияет на её стоимость**, 
что позволило бы нам "предсказать" цену на квартиру с какой-то точностью, зная только ее площадь.

### 22.2 Анализ модели

Мы помним, что нулевая гипотеза была: $H_0: b_1=0$. 
Для проверки этой гипотезы используется t-тест (результаты которого тоже есть в таблице).

В полученной выше таблице выводятся: стандартная ошибка (std err), значение t статистики и p-value ($P>|t|$).  

- В строке **hs_grad** (где $coef = -0.62 = b_1$) $p<0.05$, что позволяет нам **отклонить Н0** 
(обнаружена статистически значимая взаимосвязь между двумя нашими переменными)

    **Интерпретация $b_1$:** с каждым увеличением на **1%** жителей в штате со средним образованием число бедности будет **уменьшаться на 0.62%**.

- В строке **Intercept** (где $coef = 64.78 = b_0$) аналогично $p<0.05$. 
Здесь проверяется другая гипотеза о том, что intercept $b_0$ равен нулю ($H_0: b_0 = 0$). Как и в первом случае, **отвергаем Н0**

    **Интерпретация $b_0$:** если предположить существование такого штата, где процент среднего образования равен **нулю**,
    то мы ожидаем, что процент бедности в этом штате составит **64%**.


Значение **коэффициента детерминации** $R^2$ (R-squared), равное 0.558, означает, 
что **примерно 56% изменчивости нашей зависимой переменной объясняется нашей моделью**. 
Это не так уж и много, потому что влияние многих факторов мы не учитываем.

Построение модели и расчет всех необходимых показателей позволил нам сделать выводы, но надо также **проаналзировать остатки (residuals)**,
чтобы убедиться, что соблюдены все условия для применения **линейной регрессии*.

```python
# Calculating b0 (intercept) and b1 (slope) based on the regression model
import statsmodels.api as sm

intercept, slope = ols(formula="poverty ~ hs_grad", data=df).fit().params
X = df['hs_grad']
Y = df['poverty']

# Calculating residuals
residuals = Y - intercept - slope * X
x1 = np.linspace(min(X), max(X), len(X))

# Creating a scatterplot and a QQ-plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Scatter plot
ax1.scatter(X, residuals)
ax1.plot(x1,[0] * len(x1), '--r')

# Q-Q plot
sm.qqplot(residuals, line='s', ax = ax2)

ax1.grid()
ax1.set_title('Residuals analysis')
ax1.set_xlabel('High school education, %')
ax1.set_ylabel('Residuals')
ax1.set_xticks(np.arange(76, 93, step=4))
ax1.set_yticks(np.arange(-5, 6, step=2.5))

ax2.set_title('QQ-Plot')

fig.tight_layout(pad=3.0)
plt.show();
```

### 22.3 Анализ остатков

**Смотрим на scatter plot:**
- Видим, что у нас ~50% значений выше и ~50% значений ниже среднего + изменчивость остатков особо не меняется с ростом предиктора,
что говорит о **нормальности** распределения и о соблюдении условия **гомоскедастичности**.

**Смотрим на QQ plot:**
- Большинство остатков лежит на прямой линии, что говорит о **нормальности распределения** 
(хотя есть несколько значений, которые принимают экстремально высокие значения, т.е. несколько штатов из общей картины выпадают)

Необходимые условия соблюдены, мы можем доверять нашей модели.

## **24. Регрессионный анализ с несколькими независимыми переменными**

**Множественная регрессия (Multiple regression)** позволяет исследовать влияние сразу **нескольких** независимых переменных (НП) на **одну** зависимую (ЗП).

**Пример с двумя НП:**
Хотим исследовать, как на уровень бедности в штате (**poverty**) влияют уровень образования населения (**hs_grad**) 
и процент белого населения в штате (**white**).

Тогда уравнение модели будет иметь следующий вид:

$$ \widehat{y}=b_0+b_1X_1+b_2X_2 $$

или

$$ poverty=b_0 + b_1*\text{hs_grad} + b_2*white $$

**Уравнение модели в общем виде:**

\begin{equation*}
 \begin{cases}
   \widehat{y}=b_0+b_1X_1
   \\
   \widehat{y}=b_0+b_1X_1+b_2X_2
   \\
   ...
   \\
   \widehat{y}=b_0+b_1X_1+b_2X_2+...+b_nX_n
 \end{cases}
\end{equation*}

### 24.1 Различия с однофакторной регрессией

При использовании, к примеру, 2 предикторов (НП, факторов), вместо **регрессионной прямой** мы аппроксимируем данные **плоскостью**, 
потому что имеем **3 переменных** (2 НП + 1 ЗП), а значит - работаем в **трехмерном пространстве**.

Теперь **остатком** будет не квадрат расстояния до **регрессионной прямой**, а квадрат расстояния до **плоскости**. 

### 24.2 Метод наименьших квадратов

Аналогично с однофакторной регрессией, коэффициенты $b_0, b_1, b_2, \space ..., b_n$ подбираются так, 
чтобы **сумма всех остатков**/ошибок/квадратов расстояний от наблюдений до плоскости (или до чего-то другого, если предикторов **больше 2**) 
была **наименьшей**.

Теперь построим модель множественной регрессии и проанализируем ее


```python
df = pd.read_csv('states.csv')

# Чтобы добавить переменные, следует просуммировать в формуле поля
lm = ols(formula='poverty ~ metro_res + white + hs_grad + female_house', data=df).fit()

# Визуализируем модель для двух предикторов
'''
Построим график только для 2 предикатов
'''
def f(x, y):
    return lm.params.Intercept + lm.params.hs_grad * x  + lm.params.metro_res * y

x = df.hs_grad.sort_values()
y = df.metro_res.sort_values()

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.3)
data_below_serf = df[df['poverty'] < f(df['hs_grad'], df['metro_res'])]
data_above_serf = df[df['poverty'] >= f(df['hs_grad'], df['metro_res'])]
ax.scatter(data_below_serf.hs_grad, data_below_serf.metro_res, data_below_serf.poverty, c='r', marker='o')
ax.scatter(data_above_serf.hs_grad, data_above_serf.metro_res, data_above_serf.poverty, c='g', marker='o')
ax.set_xlabel('hs_grad')
ax.set_ylabel('metro_res')
ax.set_zlabel('poverty')

lm.summary()
```

### 25.1 Мультиколлинеарность
**Мультиколлинеарность (multicollinearity)** - сильная взаимосвязь (например, корреляция) между независимыми переменными (предикторами) 
либо идентичность переменных (например, один предиктор - рост в метрах, другой - рост в сантиментрах).

**Пример:** мы хотим предсказать успешность студентов на 1 курсе математического факультета, основываясь на результатах по ЕГЭ.
- В одном случае мы используем только один предиктор - **результат по ЕГЭ по математике**
- В другом используем два предиктора - **результат по ЕГЭ по математике** и **результат по ЕГЭ по информатике**

При этом во втором случае модель **может получиться только хуже** ($R^2$ будет меньше, чем в 1 случае), 
потому что часто у тех, у кого хорошо с математикой - хорошо и с информатикой. **Больше - НЕ значит лучше!**

**Почему так?**
1. Это просто нелогично - зачем использовать две переменных, если одна из них уже хорошо объясняет зависимую переменную?
2. Это усложняет вычисление коэффициентов регрессии с математической точки зренияъ

**Идеальная ситуация** - это когда предиктор имеет сильную взаимосвязь с ЗП, но **не** с другими предикторами.

Проверим все возможные комбинации переменных для максимизации $R^2$:

```python
# Testing all features combinations to maximize R^2

# Найдём такой набор фичей, что R-squared максимальна
features = {'metro_res', 'white', 'hs_grad', 'female_house'}

# Рекурсивно переберём все сочетания переменных
def brute_force(params: set, formula: str, result: set) -> str:
    if not params:
        result.add(formula.lstrip('+ '))
        return 
    else:
        new_params = params.copy()
        for el in params:
            new_params.remove(el)

            brute_force(new_params, formula + ' + ' + el, result)
            brute_force(new_params, formula, result)

res = set()          
brute_force(features, '', res)
res.remove('')

# Теперь посчитаем Adjusted R-Square для каждого сочетания и выведем на экран
for formula in res:
    lm = ols(formula='poverty ~ ' + formula, data=df).fit()
    print(f'{lm.rsquared_adj:.2}', formula)
```

```python
# Final model with the highest R^2

final_model = ols(formula='poverty ~ white + metro_res + hs_grad', data=df).fit()
final_model.summary()
```

Теперь проверим **нормальность распределения остатков**:

```python
# Predicted values
Intercept, white, metro_res, hs_grad = final_model.params

# Final regression model equation
y = Intercept + metro_res * df['metro_res'] + white * df['white'] + hs_grad * df['hs_grad'] 

# Getting the residuals
residuals = ols(formula="poverty ~ metro_res + white + hs_grad", data=df).fit().resid
# residuals = df['poverty'] - y - то же самое

# Plotting
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

# Scatter plot
plt.sca(ax1)
sns.scatterplot(x=y, y=residuals)
sns.lineplot(x=y, y=0, linestyle='--', color='r')
plt.title('Residuals Analysis')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')

# QQ-plot
plt.sca(ax2)
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('QQ-plot')
;
```

### 26.1 Задача классификации

**Примеры классификации:**

### 26.2 Логистическая регрессия

**Логистическая регрессия (logistic regression, logit)** - это регрессионная модель, которая позволяет исследовать взаимосвязь между **номинативной** зависимой переменной, 
имеющей **2 градации** (да/нет, 1/0), и независимыми переменными (предикторами).

**Пример:**

Пусть у нас есть данные по абитуриентам и их среднему баллу. 
- Переменная **admit** принимает значения 1 и 0 (поступил или нет)
- Переменная **mark** - это средний балл студента

С одной стороны, мы можем свести эту задачу к **сравнению средних баллов студентов в двух группах**,
т.е. ответить на вопрос "**есть ли статистически значимые различия между средним баллом поступивших и не поступивших студентов?**"

Так решение задачи выглядит в коде:


```python
admit = [0,1,1,1,0,1,1,0,1,0,0,0,1]
mark = [3.61,3.67,4,3.19,2.93,3,2.98,3.08,3.39,3.92,4,3.22,4]

df = pd.DataFrame({'admit': admit, 'mark': mark})

import seaborn as sns

ax = plt.figure(figsize=(5,3))
ax = sns.boxplot(data=df, x='admit', y='mark')

# If we want to COMPARE two populations based on our samples - 
# we could just divide the dataset into two samples (one for those who admitted and one for those who didn't) and compare them with t-test:

from scipy.stats import ttest_ind

marks_admitted = df[df.admit == 1].mark.to_list()
marks_not_admitted = df[df.admit == 0].mark.to_list()
alpha = 0.05

p_value = ttest_ind(marks_admitted, marks_not_admitted, alternative="two-sided").pvalue

if p_value > alpha:
    print("Can't Reject H0")
else:
    print("Reject H0")
```
